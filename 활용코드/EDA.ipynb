# 연극 <스크루테이프> 23.03.24 ~ 06.25
PF437124

```python
import glob
import pandas as pd

# 파일 경로 설정
file_paths = [
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_03_19_25_646,240.xlsx',
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_04_09_16_901,263.xlsx',
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_04_17_25_744,287.xlsx',
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_05_27_31_827,706.xlsx',
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_05_14_20_868,253.xlsx',
    '/content/drive/MyDrive/KOPIS공모전/데이터/(전처리)23년/2023_06_24_30_896,027.xlsx'

]

# 필터링할 공연코드
target_performance_code = 'PF437124'

# 필터링된 데이터 저장 리스트
all_filtered_data = []

# 파일 읽기 및 필터링
for file_path in file_paths:
    try:
        # 엑셀 파일 읽기
        print(f"Reading file: {file_path}")
        df = pd.read_excel(file_path, engine='openpyxl')

        # 공연코드가 PF301099인 데이터 필터링
        filtered_data = df[df['공연코드'] == target_performance_code]

        # 필터링된 데이터가 비어 있지 않은지 확인
        if not filtered_data.empty:
            all_filtered_data.append(filtered_data)

    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# 필터링된 데이터 병합
data1 = pd.concat(all_filtered_data, ignore_index=True)

# 결과 출력
data1.head()
```

```python
data1['재관람'].value_counts()
```

```python
# 날짜 및 시간 칼럼 처리 (errors='coerce'로 잘못된 날짜는 NaT로 처리)
data1['공연일시'] = pd.to_datetime(data1['공연일시'], errors='coerce')
data1['예매/취소일시'] = pd.to_datetime(data1['예매/취소일시'], errors='coerce')
data1['판매시작일시'] = pd.to_datetime(data1['판매시작일시'], errors='coerce')

# 예매/취소매수 계산
booking_counts = data1.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소매수'].sum().unstack(fill_value=0)
booking_counts['총 예매/취소매수'] = booking_counts.get(1, 0) - booking_counts.get(2, 0)

# 예매/취소금액 계산
booking_amounts = data1.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소금액'].sum().unstack(fill_value=0)
booking_amounts['총 예매/취소금액'] = booking_amounts.get(1, 0) - booking_amounts.get(2, 0)

# 할인금액 계산
discount_amounts = data1.groupby(['공연코드', '공연일시', '예매/취소구분'])['할인금액'].sum().unstack(fill_value=0)
discount_amounts['총 할인금액'] = discount_amounts.get(1, 0) - discount_amounts.get(2, 0)

# 각 항목을 병합
result = pd.merge(booking_counts[['총 예매/취소매수']].reset_index(),
                  booking_amounts[['총 예매/취소금액']].reset_index(),
                  on=['공연코드', '공연일시'])

result = pd.merge(result, discount_amounts[['총 할인금액']].reset_index(),
                  on=['공연코드', '공연일시'])

# 좌석수 컬럼이 포함된 데이터로 좌석수 가져오기
seat_counts = data1.groupby(['공연코드', '공연일시'])['좌석수'].first().reset_index()

# 채움 비율 계산을 위한 데이터 병합
performance_group1 = pd.merge(result, seat_counts, on=['공연코드', '공연일시'])

# 채움 비율 계산
performance_group1['채움비율'] = performance_group1['총 예매/취소매수'] / performance_group1['좌석수']

# 결과 출력
performance_group1 = performance_group1[['공연코드', '공연일시', '총 예매/취소매수', '총 예매/취소금액', '총 할인금액', '좌석수', '채움비율']]
performance_group1
```

```python
#(중간일자 채움비율 평균) / ((첫공 채움비율 + 막공 채움비율)/2)
# 첫 공연과 막 공연의 채움비율 평균 계산
first_last_avg1 = (performance_group1['채움비율'].iloc[-1:]) / 2

# 중간일자의 채움비율 평균 계산 (첫 공연과 막 공연 제외)
middle_fill_rate_avg1 = performance_group1['채움비율'].iloc[1:-1].mean()

# 계산 결과
result1 = middle_fill_rate_avg1 / first_last_avg1
result1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 첫 공연, 중간 평균, 막 공연의 채움비율 계산
first_fill_rate1 = performance_group1['채움비율'].iloc[0]
last_fill_rate2 = performance_group1['채움비율'].iloc[-1]
middle_fill_rate_avg2 = performance_group1['채움비율'].iloc[1:-1].mean()

# 데이터프레임 생성 (시각화를 위한 데이터)
visualization_df3 = pd.DataFrame({
    'Category': ['First Performance', 'Middle Average', 'Last Performance'],
    'Fill Rate': [first_fill_rate1, middle_fill_rate_avg2, last_fill_rate2]
})

# 시각화
plt.figure(figsize=(8, 6))
sns.lineplot(x='Category', y='Fill Rate', data=visualization_df3, marker='o', linewidth=2.5, color='b')
plt.ylim(0, 1)  # 채움비율이 0~1 사이이므로 y축 범위를 설정
plt.title('Fill Rate Comparison: First, Middle Average, Last Performance')
plt.ylabel('Fill Rate')
plt.xlabel('Performance Category')
plt.show()
```

```python
# 날짜 및 시간 칼럼 처리 (errors='coerce'로 잘못된 날짜는 NaT로 처리)
data3['공연일시'] = pd.to_datetime(data3['공연일시'], errors='coerce')
data3['예매/취소일시'] = pd.to_datetime(data3['예매/취소일시'], errors='coerce')
data3['판매시작일시'] = pd.to_datetime(data3['판매시작일시'], errors='coerce')

# 예매/취소매수 계산
booking_counts = data3.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소매수'].sum().unstack(fill_value=0)
booking_counts['총 예매/취소매수'] = booking_counts.get(1, 0) - booking_counts.get(2, 0)

# 예매/취소금액 계산
booking_amounts = data3.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소금액'].sum().unstack(fill_value=0)
booking_amounts['총 예매/취소금액'] = booking_amounts.get(1, 0) - booking_amounts.get(2, 0)

# 할인금액 계산
discount_amounts = data3.groupby(['공연코드', '공연일시', '예매/취소구분'])['할인금액'].sum().unstack(fill_value=0)
discount_amounts['총 할인금액'] = discount_amounts.get(1, 0) - discount_amounts.get(2, 0)

# 각 항목을 병합
result = pd.merge(booking_counts[['총 예매/취소매수']].reset_index(),
                  booking_amounts[['총 예매/취소금액']].reset_index(),
                  on=['공연코드', '공연일시'])

result = pd.merge(result, discount_amounts[['총 할인금액']].reset_index(),
                  on=['공연코드', '공연일시'])

# 좌석수 컬럼이 포함된 데이터로 좌석수 가져오기
seat_counts = data3.groupby(['공연코드', '공연일시'])['좌석수'].first().reset_index()

# 채움 비율 계산을 위한 데이터 병합
performance_group3 = pd.merge(result, seat_counts, on=['공연코드', '공연일시'])

# 채움 비율 계산
performance_group3['채움비율'] = performance_group3['총 예매/취소매수'] / performance_group3['좌석수']

# 결과 출력
performance_group3 = performance_group3[['공연코드', '공연일시', '총 예매/취소매수', '총 예매/취소금액', '총 할인금액', '좌석수', '채움비율']]
performance_group3
```

```python
#(중간일자 채움비율 평균) / ((첫공 채움비율 + 막공 채움비율)/2)
# 첫 공연과 막 공연의 채움비율 평균 계산
first_last_avg3 = (performance_group3['채움비율'].iloc[0:6].mean() + performance_group3['채움비율'].iloc[-3].mean()) / 2

# 중간일자의 채움비율 평균 계산 (첫 공연과 막 공연 제외)
middle_fill_rate_avg3 = performance_group3['채움비율'].iloc[5:-3].mean()

# 계산 결과
result3 = middle_fill_rate_avg3 / first_last_avg3
result3
```

```python
# 날짜 및 시간 칼럼 처리 (errors='coerce'로 잘못된 날짜는 NaT로 처리)
data9['공연일시'] = pd.to_datetime(data9['공연일시'], errors='coerce')
data9['예매/취소일시'] = pd.to_datetime(data9['예매/취소일시'], errors='coerce')
data9['판매시작일시'] = pd.to_datetime(data9['판매시작일시'], errors='coerce')

# 예매/취소매수 계산
booking_counts = data9.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소매수'].sum().unstack(fill_value=0)
booking_counts['총 예매/취소매수'] = booking_counts.get(1, 0) - booking_counts.get(2, 0)

# 예매/취소금액 계산
booking_amounts = data9.groupby(['공연코드', '공연일시', '예매/취소구분'])['예매/취소금액'].sum().unstack(fill_value=0)
booking_amounts['총 예매/취소금액'] = booking_amounts.get(1, 0) - booking_amounts.get(2, 0)

# 할인금액 계산
discount_amounts = data9.groupby(['공연코드', '공연일시', '예매/취소구분'])['할인금액'].sum().unstack(fill_value=0)
discount_amounts['총 할인금액'] = discount_amounts.get(1, 0) - discount_amounts.get(2, 0)

# 각 항목을 병합
result = pd.merge(booking_counts[['총 예매/취소매수']].reset_index(),
                  booking_amounts[['총 예매/취소금액']].reset_index(),
                  on=['공연코드', '공연일시'])

result = pd.merge(result, discount_amounts[['총 할인금액']].reset_index(),
                  on=['공연코드', '공연일시'])

# 좌석수 컬럼이 포함된 데이터로 좌석수 가져오기
seat_counts = data9.groupby(['공연코드', '공연일시'])['좌석수'].first().reset_index()

# 채움 비율 계산을 위한 데이터 병합
performance_group9 = pd.merge(result, seat_counts, on=['공연코드', '공연일시'])

# 채움 비율 계산
performance_group9['채움비율'] = performance_group9['총 예매/취소매수'] / performance_group9['좌석수']

# 결과 출력
performance_group9 = performance_group9[['공연코드', '공연일시', '총 예매/취소매수', '총 예매/취소금액', '총 할인금액', '좌석수', '채움비율']]
performance_group9
```

```python
#(중간일자 채움비율 평균) / ((첫공 채움비율 + 막공 채움비율)/2)

# 첫 공연과 막 공연의 채움비율 평균 계산
first_last_avg9 = performance_group9['채움비율'].iloc[-4].mean()

# 중간일자의 채움비율 평균 계산 (첫 공연과 막 공연 제외)
middle_fill_rate_avg9 = performance_group9['채움비율'].iloc[0:-4].mean()

# 계산 결과
result9 = middle_fill_rate_avg9 / first_last_avg9
result9
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 첫 공연, 중간 평균, 막 공연의 채움비율 계산
last_fill_rate9 = performance_group9['채움비율'].iloc[-1]
middle_fill_rate_avg9 = performance_group9['채움비율'].iloc[3:-3].mean()

# 데이터프레임 생성 (시각화를 위한 데이터)
visualization_df = pd.DataFrame({
    'Category': [ 'Middle Average', 'Last Performance'],
    'Fill Rate': [middle_fill_rate_avg9, last_fill_rate9]
})

# 시각화
plt.figure(figsize=(8, 6))
sns.lineplot(x='Category', y='Fill Rate', data=visualization_df, marker='o', linewidth=2.5, color='b')
plt.ylim(0, 1)  # 채움비율이 0~1 사이이므로 y축 범위를 설정
plt.title('Fill Rate Comparison: Middle Average, Last Performance')
plt.ylabel('Fill Rate')
plt.xlabel('Performance Category : School of Rock(Musical)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 특정 공연코드만 시각화
filtered_data9 = performance_group9[performance_group9['공연코드'] == 'PF369984']

# 데이터프레임이 정렬되어 있다고 가정합니다. 만약 그렇지 않다면 먼저 정렬이 필요합니다.
filtered_data9 = filtered_data9.sort_values(by='공연일시')

# 앞의 3개, 뒤의 3개 데이터 고정
fixed_data = pd.concat([filtered_data9.head(3), filtered_data9.tail(3)])

# 중간 데이터에서 랜덤하게 7개 선택
middle_data = filtered_data9.iloc[3:-3]

# 고정된 데이터와 랜덤 선택된 중간 데이터를 합침
combined_data = pd.concat([fixed_data, middle_data])

# 다시 날짜 순으로 정렬
combined_data = combined_data.sort_values(by='공연일시')

# U자형 패턴 확인 - 채움 비율 시각화
plt.figure(figsize=(12, 6))
plt.scatter(combined_data['공연일시'], combined_data['채움비율'], color='blue', marker='o', label='Sampled Data')

# 추세선을 그리기 위해 다항회귀를 사용하여 선형성을 추정합니다.
z = np.polyfit(pd.to_datetime(combined_data['공연일시']).astype(int), combined_data['채움비율'], 2)
p = np.poly1d(z)

# 공연일시를 숫자로 변환하여 추세선 그리기
plt.plot(combined_data['공연일시'], p(pd.to_datetime(combined_data['공연일시']).astype(int)), "r--", label='Trend Line')

plt.xlabel('Performance Date')
plt.ylabel('Occupancy rate')
plt.title('Musical & Occupancy rate : School of Rock(Musical)')
plt.grid(True)
plt.legend()
plt.show()
```
