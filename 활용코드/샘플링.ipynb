```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
def stratified_sampling(df, genre_col, size_col, sample_size=0.5):
    df['계층_조합'] = df[genre_col].astype(str) + '_' + df[size_col].astype(str)
    stratified_sample, _ = train_test_split(
        df, 
        test_size=1-sample_size, 
        stratify=df['계층_조합']
    )
    
    return stratified_sample
```

```python
for filename in os.listdir():
    if filename.endswith('.xlsx'):
        try:
            # 파일을 읽어보는 시도
            d = pd.read_excel(filename)
        except Exception as e:
            # 에러가 발생하면 파일 이름과 에러 메시지를 출력
            print(f"Error reading {filename}: {e}")
```

```python
df = pd.DataFrame()

for filename in os.listdir() :
    if filename.endswith('.xlsx') :
        d = pd.read_excel(filename)
        sampled_df = stratified_sampling(d, genre_col = '장르명', size_col = '극장규모')
        df = pd.concat([df, sampled_df], ignore_index=True)
```
